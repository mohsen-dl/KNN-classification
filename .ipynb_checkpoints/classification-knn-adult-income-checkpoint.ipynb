education-num: continuous.

occupation: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces.

## standardizes features

```python
sc = preprocessing.StandardScaler().fit(X)
X = sc.transform(X.astype(float))
X[0:5]
```

## splitting data for train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
print('train set : ', x_train.shape, y_train.shape)
print('test set : ', x_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(x_train)
x_train_sc = scaler.transform(x_train)

x_test_sc = scaler.transform(x_test)
x_train_sc[0:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=4
neigh = KNeighborsClassifier(n_neighbors=k).fit(x_train_sc, y_train)
neigh
```

```python
yhat = neigh.predict(x_test_sc)
yhat[0:5]
```

```python
from sklearn import metrics
```

## checking accuracy

```python
score = metrics.accuracy_score(y_test, yhat)
print('train set accuracy: ', metrics.accuracy_score(y_train, neigh.predict(x_train_sc)))
print('test set accuracy: ', score)
```

```python
ks=12
mean_acc = np.zeros(ks-1)
std_acc = np.zeros(ks-1)
k_values = np.zeros(ks-1)

for n in range(1,ks):
    k_values[n-1] = n
    neigh = KNeighborsClassifier(n_neighbors=n).fit(x_train_sc, y_train)
    yhat = neigh.predict(x_test_sc)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)
    
    std_acc[n-1] = np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
    
mean_acc
```

```python
std_acc
```

```python
plt.plot(k_values, mean_acc)
plt.title('accuracy score')
plt.xlabel('k values')
plt.ylabel('accuracy score for diffrent k')
plt.show()
```

## confusion_matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test,yhat)
cm
```

```python
cm = confusion_matrix(y_test, yhat, labels=neigh.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=neigh.classes_)
disp.plot()

plt.show()
```

## resampling

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
x_sm, y_sm = smote.fit_resample(x_train, y_train)
```

```python
scaler = StandardScaler().fit(x_sm)
x_train_sm_sc = scaler.transform(x_sm)

x_train_sm_sc[0:5]
```
